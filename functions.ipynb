```python
# *args, **kwargs 

def add(*args):
    total = 0 
    for i in args:
        total+= i
    return total
print(f'Adding with different numbers of inputs: {add(1,2,3),add(1,2)}')

def user(**kwargs):
    for k,w in kwargs.items():
        print(k,':',w)
print('\nUsing different named inputs:')
user(name='eps',age=25)
print('-----')
user(name='jake',age=23,height=180)
```

```python
# Closures 
def add5(x):
    def add(y):
        return x+y
    return add 


add = add5(5)
print(f'Function that will remember 5 and add: {add(3),add(4),add(5)}')
```
