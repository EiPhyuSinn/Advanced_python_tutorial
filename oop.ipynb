```python
class Person:
    def __init__(self,name):
        self.name = name

p1 = Person('eps')
print(p1)

# Using __str__
class Person:
    def __init__(self,name):
        self.name = name

    def __str__(self):
        return f'Name is {self.name}'
    
p2 = Person('eps')
print(p2)

# Using __repr__
class Person:
    def __init__(self, name):
        self.name = name
    
    def __repr__(self):
        return f"Person(name='{self.name}')"
    
p3 = Person('eps')
print(p3)
```

```python
# @classmethod 

class Dog:
    species = 'Pug' # class variable -> belongs to class,same for every object 
    count = 0
    names = []
    
    def __init__(self,name):
        self.name = name # instance variable -> belongs to one specific object
        Dog.count +=1 
        Dog.names.append(self.name)

    @classmethod
    def showcount(cls):
        return cls.count

    @classmethod
    def showDogs(cls):
        return cls.names
        

d1 = Dog('Pae Pae')
d1.species #Pug
d2 = Dog('Kae Kae')
d2.species #Pug

print(f"Total dog counts: {Dog.showcount()}")
print(f"Total dogs : {Dog.showDogs()}")
```

```python
# staticmethod 

class Person:
    def greet(name):
        print(f'Hello I am {name}')

Person.greet('eps') #This will work because it is a normal function without self 
p = Person('eps') 
p.greet() #This will fail -> Person() takes no arguments

class Person:
    @staticmethod
    def greet(name):
        return f'Hello I am {name}'

p = Person()
p.greet('eps')  # Hello I am eps
Person.greet('eps')  # Hello I am eps
```

```python
# Inheritance and Composition

class Animal:
    def makesound(self,sound):
        return f'{self.name} says {sound}' 
    
class Dog(Animal):
    def __init__(self,name):
        self.name = name 

d = Dog('Kae Kae')
d.makesound('Woof!')

class Owner:
    def __init__(self,name,dog_name):
        self.name = name 
        self.dog_name = dog_name
        self.dog = Dog(dog_name)

    def statement(self):
        print(f'{self.name} has a dog named {self.dog_name}')
        return self.dog.makesound('woof!')
    

p = Owner('Eps','Pae Pae')
p.statement()
```
