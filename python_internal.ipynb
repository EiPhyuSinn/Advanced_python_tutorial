```python
# Mutable Vs Immutable 

# Immutable -> cannot change, if you change the value, python creates new object (int, str, float, tuple)
a = 3
b = 3 
print(f'Address of a,b,3 : {id(a),id(b),id(3)}') # Python variables do not store values, they store references(addresses), same address for same value

# Mutable -> can change, list, dict, set, bytearray
a = [1,2,3]
print(f'Address of a : {id(a)}') 
a.append(4) # same object, dont change reference 
print(f'Address of a after append : {id(a)}') 
```

```python
# Shallow Copy 
import copy
original = [1, 2, [3, 4]]
shallow_copy = copy.copy(original)
shallow_copy.append(5)
print(f'shallow_copy : {shallow_copy} -> original -> {original}')
shallow_copy[2].append(5)
print(f'shallow_copy : {shallow_copy} -> original -> {original}')


# Deep Copy
original = [1, 2, [3, 4]]
deep_copy = copy.deepcopy(original)
deep_copy.append(5)
print(f'deep_copy : {deep_copy} -> original -> {original}')
deep_copy[2].append(5)
print(f'deep_copy : {deep_copy} -> original -> {original}')
```
