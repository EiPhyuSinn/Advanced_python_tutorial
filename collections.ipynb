```python
# Default Dict 
from collections import defaultdict

d = defaultdict(int)
print(f'defaultdict of int : {d['a']}')

d = defaultdict(str)
print(f'defaultdict of str : {d['a']}')

d = defaultdict(float)
print(f'defaultdict of float : {d['a']}')

d = defaultdict(list)
print(f'defaultdict of list : {d['a']}')

d = defaultdict(set)
print(f'defaultdict of set : {d['a']}')

d = defaultdict(dict)
print(f'defaultdict of dict : {d['a']}')

def minusOne():
    return -1
d = defaultdict(minusOne)
print(f'defaultdict of own function : {d['a']}')

# Example usage of Defaultdict 
word = 'banana'
counts = defaultdict(int)
for w in word:
    counts[w]+= 1 
print(f'\nWord counts in {word} : {counts}')
```

```python
# Counter
from collections import Counter

print(f'Methods in Counter : {dir(Counter())}')

word = 'banana'
counts = Counter(word)
print(f'Word counts using Counter in {word} : {counts}')
print(f'Using __add__ method to add more counts : {counts.__add__(Counter("eps"))}')
print(f'Most common letters : {counts.most_common(2)}')
```

```python
# Deque
from collections import deque

q = deque()
print(f'Methods in Deque : {[m for m in dir(q) if not m.startswith("__")]}')

# Normal List 
li = [1,2,3]
li.append(1)
print(f'li after append : {li}')
li.pop(1)
print(f'li after pop : {li}')

# Using Deque 
q = deque([1,2,3])
q.appendleft(0),q.append(4)
print(f'Appending both : {q}')
q.popleft(),q.pop()
print(f'Appending both : {q}')
```

```python
# Namedtuple 
from collections import namedtuple

Student = namedtuple('student',['name','age'])

s = Student('eps',25)
print(f'Student name:{s.name}, age:{s.age}')
```

```python
# Dataclass 

class Person: #Normal class 
    def __init__(self,name,age):
        self.name = name 
        self.age = age 
    def __str__(self):
        return f'Name : {self.name}, Age : {self.age}'
    def __eq__(self, other):
        return self.name == other.name 

p1 = Person('eps',25)
p2 = Person('jake',23)
p3 = Person('eps',23)
print(p1 == p3)
print(p1 == p2)

from dataclasses import dataclass
@dataclass
class Person:
    name : str
    age : int 

p1 = Person('eps',25)
p2 = Person('jake',23)
p3 = Person('eps',25)
print(p1 == p3)
print(p1 == p2)
```
