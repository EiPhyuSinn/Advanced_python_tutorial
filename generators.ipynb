```python
# iter() -> An iterator gives values one at a time, only when you ask. 

nums = [1, 2, 3]
it = iter(nums)
while True:
    try:
        print(next(it))
    except StopIteration:
        break
```

```python
# Generator function -> A generator function is defined like a normal function, but whenever it needs to generate a value, it does so with the yield keyword rather than return. If the body of a function contains yield, the function automatically becomes a generator function.
def traffic_light():
    while True:
        yield "Red"
        yield "Yellow"
        yield "Green"
        
light = traffic_light()
print(next(light))  # Red
print(next(light))  # Yellow
print(next(light))  # Green
print(next(light))  # Red
print(next(light))  # Yellow


for i in range(3):
    print(traffic_light())
```

```python
# Generator comprehension -> Similar to list comprehensions, but using parentheses instead of square brackets.
import sys

list_squares = [x * x for x in range(100000)]
gen_squares = (x * x for x in range(100000))

print("List memory:", sys.getsizeof(list_squares))
print("Generator memory:", sys.getsizeof(gen_squares))

print("List memory of each index:", sys.getsizeof(list_squares[0]))
print("Generator memory of each index:", sys.getsizeof(next(gen_squares)))
```
